```python
import random
import matplotlib.pyplot as plt
# import numpy as np


def FS(S):
    if S>0 :
        return 1
    else:
        return -1

random.seed()

dane = [
    [2.0, 1.0,  1.0],
    [2.0, 2.0, 1.0],
    [0.0, 6.0, 1.0],
    [-2.0, 10.0, -1.0],
    [-2.0, 0.0, -1.0],
    [0.0, 0.0, -1.0],
    [4.0,-20.0,-1.0]
]
wagi = [0.0,0.0,0.0]
x0 = 1

# iteracje = 100 
iteracje = 0

print(f"Poczatkowe wagi: {wagi}")




# for it in range(iteracje):
while True:
    iteracje+=1
    pomylki = 0
    #print(f"\n--- Iteracja {it+1} ---")
    print(f"\nIteracja {iteracje}:")
    for i in range(len(dane)):
        x1 = dane[i][0]
        x2 = dane[i][1]
        d = dane[i][2]
        y_in = wagi[0]*x0 + wagi[1]*x1 + wagi[2]*x2
        y = FS(y_in)
        print(f" Przetwarzany wektor: [{x0}, {x1}, {x2}] z oczekiwana wartoscia {d} => y_in = {y_in} => y = {y}")
        # if((wagi[0]*x0+x1*wagi[1]+x2*wagi[2])!=0):
            # pass

        if y !=d :
            wagi[0] = wagi[0] + d*x0
            wagi[1] = wagi[1] + d*x1
            wagi[2] = wagi[2] + d*x2
            print(f" Bledna klasyfikacja -> aktualizacja wag: {wagi}")
            pomylki += 1
    print(f"\nDane: {dane}\n Wagi: {wagi}\n Liczba bledow w tej iteracji: {pomylki}")
    if pomylki == 0:
        print(" Brak bledow, zakonczono nauke.")
        break
print(f"\nKoncowe wagi: {wagi}")
plt.ion()
plt.figure()
plt.xlim(-15,15)
plt.ylim(-25,15)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True)
plt.title("Perceptron Learning Algorithm")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter([point[0] for point in dane if point[2] == 1],
            [point[1] for point in dane if point[2] == 1],
            color='blue', marker='o', label='Class 1')
plt.scatter([point[0] for point in dane if point[2] == -1],
            [point[1] for point in dane if point[2] == -1],
            color='red', marker='x', label='Class -1')
plt.legend()
plt.draw()
plt.pause(0.1)
for i in dane:
    x1,x2,d=i
    y_in = wagi[0]*x0 + wagi[1]*x1 + wagi[2]*x2
    y = FS(y_in)
    print(f" Wejscie: [{x0}, {x1}, {x2}] oczekiwana wartosc: {d} => y_in = {y_in} => y = {y}")
```
